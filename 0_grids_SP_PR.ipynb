## Bounding Box

```python
from typing import Union
import geopandas as gpd
from shapely.geometry import Polygon

def points_to_bounding_box(points: Union[gpd.GeoSeries, gpd.GeoDataFrame]):
    """
    Function transform a set of points into a Polygon with bounding box.
    
    Parameters
    ----------
    points : gpd.GeoSeries
    
    Returns
    -------
    bounding_box : Polygon
    """
    bounds = points.total_bounds
    
    x_min, y_min, x_max, y_max = bounds
    
    polygon_tuples = [
        (x_min, y_min), (x_max, y_min), (x_max, y_max), (x_min, y_max)
    ]
    
    polygon = Polygon(polygon_tuples)
    return polygon

centroids = gpd.GeoDataFrame(df_final, crs = df_final.crs)
poly = points_to_bounding_box(centroids)
print(poly)
```

```python
# total area for the grid
xmin, ymin, xmax, ymax= centroids.total_bounds
# how many cells across and down
n_cells=160
cell_size = (xmax-xmin)/n_cells
print(cell_size)
# create the cells in a loop
grid_cells = []
for x0 in np.arange(xmin, xmax+cell_size, cell_size ):
    for y0 in np.arange(ymin, ymax+cell_size, cell_size):
        # bounds
        x1 = x0-cell_size
        y1 = y0+cell_size
        grid_cells.append(box(x0, y0, x1, y1))

# projection of the grid ### TO DO: put the right projection
cell = gpd.GeoDataFrame(grid_cells, columns=['geometry'], crs=centroids.crs)
```

```python
def add_values_in_dict(sample_dict, key, list_of_values):
    ''' Append multiple values to a key in 
        the given dictionary '''
    if key not in sample_dict:
        sample_dict[key] = list()
    sample_dict[key].extend(list_of_values)
    return sample_dict
```

```python
def weighted_average(merged, indicator):
  dic = {}
  # p cada index encontre o setores
  for i in range(len(merged)):
    final_dic = add_values_in_dict(dic, merged['cell_id'][i], [merged['percentage'][i] , merged[indicator][i]])
  for key, list_of_values in final_dic.items():
    num = 0
    den = 0
    number_of_elements = len(list_of_values)
    for i in range(number_of_elements-1):
      if i % 2 == 0:
        num += list_of_values[i] * list_of_values[i+1] # percentage * value
        den += list_of_values[i] # sum of percentages
    final_dic[key] = num / den
  
  return final_dic
```

```python
def dataframe(merged, indicator):
  final_dic = weighted_average(merged, indicator)
  df_weighted_average = pd.DataFrame.from_dict(final_dic, orient='index')
  df_weighted_average.rename(columns = {0:indicator}, inplace = True)
  df_weighted_average['id'] = 1
  for i in range(len(df_weighted_average)):
    df_weighted_average['id'][i] = i
  return df_weighted_average
```

```python
df_weighted_average_idh = dataframe(merged, 'IDHM')
df_weighted_average_income = dataframe(merged, 'IncomeIDHM')
df_weighted_average_literacy = dataframe(merged, 'LiteracyIDHM')
df_weighted_average_longevity = dataframe(merged, 'LongevityIDHM')
```

## Division in groups

```python
import math
```

```python
n = math.floor((len(final_cell_3857)/30))
n
```

```python
N = 30 #number of groups 
n = math.floor((len(final_cell_3857)/N))
final_cell_3857['group'] = 1

def group(k, n):
  if k <= N-1:
    for i in range((k-1)*n, k*n):
      final_cell_3857['group'][i] = 'group{k}'.format(k=k)
  else:
    for i in range((k-1)*n, len(final_cell_3857)):
      final_cell_3857['group'][i] = 'group{k}'.format(k=k)


for k in range(1,N+1):
  group(k, n)


final_cell_3857
```
